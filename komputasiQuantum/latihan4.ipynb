```python
# membuat pogram menebak bilangan 
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
#variabel numbernya
number_saya = '11100100010010110'
```

```python
#membangun sircuit
circuit = QuantumCircuit(len(number_saya)+1,len(number_saya))
#circuit.h([0,1,2,3,4,5])
circuit.h(range(len(number_saya)))
circuit.x(len(number_saya))
circuit.h(len(number_saya))

#membuat batasan
circuit.barrier()

for ii, yesno in enumerate(reversed(number_saya)):
    if yesno == '1':
        circuit.cx(ii,len(number_saya))
#circuit.cx(5,6)
#circuit.cx(3,6)
#circuit.cx(0,6)

#membuat batasan
circuit.barrier()
#circuit.h([0,1,2,3,4,5])
circuit.h(range(len(number_saya)))
circuit.barrier()
circuit.measure(range(len(number_saya)), range(len(number_saya)))
```

```python
circuit.draw()
```

```python
#membuktikan jalan atau tidak untuk menebak
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, simulator,shots=1).result()
counts = result.get_counts()
print(counts)
```
