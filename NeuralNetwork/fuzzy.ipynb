```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
def rentang(x1,x2):
    x = np.arange(x1,x2,0.1)
    xn = (x1+x2)/2
    y = []
    for i in x:
        if i <= x1 or i >= x2:
            y.append(0)
        elif i >= x1 and i <= xn:
            y.append((i-x1)/(xn-x1))
        else:
            y.append((x2-i)/(x2-xn))
    return x,np.array(y)

x1,y1 = rentang(10,20)
x2,y2 = rentang(15,25)
x3,y3 = rentang(20,30)
x4,y4 = rentang(25,35)
x5,y5 = rentang(30,40)
x6,y6 = rentang(35,45)

fig,ax = plt.subplots(figsize=(10,5))
ax.plot(x1,y1,label="Rentang 1")
ax.plot(x2,y2,label="Rentang 2")
ax.plot(x3,y3,label="Rentang 3")
ax.plot(x4,y4,label="Rentang 4")
ax.plot(x5,y5,label="Rentang 5")
ax.plot(x6,y6,label="Rentang 6")
plt.title("Rentang Fuzzy")
plt.xlabel("Nilai")
plt.ylabel("Nilai Fuzzy")
plt.legend()
plt.axis([15,40,0,1])
plt.grid(True)
plt.show()
```
