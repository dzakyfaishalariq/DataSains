```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
import seaborn as sns
```

```python
#Konversi ke citra biner
citra = cv2.imread('img/gambarkeluarga.jpeg',cv2.IMREAD_GRAYSCALE)
ambang, binerA = cv2.threshold(citra,127,255,cv2.THRESH_BINARY)
ambang, binerB = cv2.threshold(citra,127,255,cv2.THRESH_BINARY_INV)

#mengubah ke RGB color
rgb1 = citra[...,::-1]
rgb2 = binerA[...,::-1]
rgb3 = binerB[...,::-1]
#tampilkan ke matloblib
fig, ax = plt.subplots(1,3)
ax[0].imshow(rgb1)
ax[0].set_title('Gambar Asli')
ax[1].imshow(rgb2)
ax[1].set_title('Biner A')
ax[2].imshow(rgb3)
ax[2].set_title('Biner B')

plt.show()

#cv2.imshow('Gambar',rgb1)
#cv2.imshow('Gambar Biner A',rgb2)
#cv2.imshow('Gambar Biner B',rgb3)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
citra = cv2.imread('img/gambarkeluarga.jpeg',cv2.IMREAD_GRAYSCALE)
ambang, binerA = cv2.threshold(citra,127,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

rgb1 = citra[...,::-1]
rgb2 = binerA[...,::-1]
fig, ax = plt.subplots(1,2)
ax[0].imshow(rgb1)
ax[0].set_title('Gambar Asli')
ax[1].imshow(rgb2)
ax[1].set_title('Biner A')
plt.show()
```

```python
#pembuatan garis pusat
berkas = cv2.imread('img/gambarkeluarga.jpeg',cv2.IMREAD_UNCHANGED)
jumlah_kolom = berkas.shape[1]
jumlah_baris = berkas.shape[0]
x_garis = jumlah_kolom // 2
y_garis = jumlah_baris // 2
# buat garis tegak
cv2.line
(
        berkas,
        (x_garis,0),
        (y_garis,jumlah_baris-1),
        (128,128,128),
        5
)
cv2.line
(
    berkas,
    (0,y_garis),
    (jumlah_kolom-1,y_garis),
    (128,128,128),
    5
)
#citra = berkas[...,::-1]
cv2.imshow('Gambar',berkas)
cv2.waitKey(0)
cv2.destroyAllWindows()
#plt.imshow(citra)
#plt.show()
```
